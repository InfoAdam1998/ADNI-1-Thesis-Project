```python
import os
from dotenv import load_dotenv

import pandas as pd
```

```python
load_dotenv()
dataset_path=os.getenv("RAW_ADNI_DATASET")
processed_path = os.getenv("PROCESSED_ADNI_DATASET")

dataset = pd.read_excel(dataset_path)
dataset.head()
```

```python
def adni_data_preprocessing(dataset, output_path):
    
    # Keep only rows where Diagnosis is not 2
    dataset = dataset[dataset["Diagnosis"] != 2].reset_index(drop=True)

    # Shuffle rows
    dataset = dataset.sample(frac=1, random_state=42).reset_index(drop=True)

    # Export to CSV (hardcoded path)
    dataset.to_csv(output_path,index=False)
    
    return dataset
```

```python
adni_data_preprocessing(dataset, processed_path)
```
